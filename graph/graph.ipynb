# BFS (Breadth First Search)
-            ,          

-      ,     ,     

- BFS      



-   :  G   V







```python
def bfs(node):
    q = [node]

    while q:
        now = q.pop(0)

        print(now, end=' ')  #   

        for next_node in graph[now]:
            if visited[next_node]:
                continue

            visited[next_node] = 1
            q.append(next_node)


N, M = map(int, input().split())
graph = [[] for _ in range(N + 1)]
visited = [0] * (N + 1)
for _ in range(M):
    s, e = map(int, input().split())
    graph[s].append(e)
    graph[e].append(s)

visited[1] = 1
bfs(1)
```

```python
def make_set(n):
    p = [i for i in range(n)]  #     
    r = [0] * n
    return p, r


def find(x):
    #    ==    
    if parents[x] == x:
        return x

    #   (path compression)    
    parents[x] = find(parents[x])
    return parents[x]


def union(x, y):
    root_x = find(x)
    root_y = find(y)

    if root_x == root_y:  #    
        return

    # # rank        
    if ranks[root_x] > ranks[root_y]:
        parents[root_y] = root_x
    elif ranks[root_x] < ranks[root_y]:
        parents[root_x] = root_y
    else:
        # rank       rank 
        parents[root_y] = root_x
        ranks[root_x] += 1


#  
n = 7  #  
parents, ranks = make_set(n)

union(1, 3)
union(2, 3)
union(5, 6)

print('find_set(6) = ', find(6))

target_x = 2
target_y = 3

#  1  2     
if find(target_x) == find(target_y):
    print(f" {target_x}  {target_y}    .")
else:
    print(f" {target_x}  {target_y}    .")
```
